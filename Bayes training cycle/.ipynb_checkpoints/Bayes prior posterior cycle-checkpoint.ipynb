```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define the parameters for the prior Beta distribution
alpha_prior = 2
beta_prior = 2

# Plot the prior distribution
x = np.linspace(0, 1, 100)
prior = beta.pdf(x, alpha_prior, beta_prior)

plt.plot(x, prior, label='Prior (Beta(2,2))')
plt.title('Prior Distribution')
plt.xlabel('Probability of Success')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Update the parameters with the observed data
alpha_posterior = alpha_prior + 7
beta_posterior = beta_prior + 10 - 7

# Plot the posterior distribution
posterior = beta.pdf(x, alpha_posterior, beta_posterior)

plt.plot(x, posterior, label='Posterior (Beta(9,5))')
plt.title('Posterior Distribution')
plt.xlabel('Probability of Success')
plt.ylabel('Density')
plt.legend()
plt.show()
```
