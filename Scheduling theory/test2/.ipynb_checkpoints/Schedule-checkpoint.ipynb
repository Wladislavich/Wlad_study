```python
from ortools.sat.python import cp_model
import numpy as np
import pandas as pd
```

```python
#Показатели в часах
month_hours = 175
week_hours = 40
min_day_time = 4
max_day_time = 12
#Показатели, которые не будут корректироваться
min_break_cnt = 0
max_break_cnt = 2
max_work_days = 6
```

```python
#Загрузка исходных данных и инициализация модели
data = pd.read_csv('template.csv', sep=';')
model = cp_model.CpModel()
```

```python
month_hours *= 2
week_hours *= 2
min_day_time *=2
max_day_time *=2

num_employee = round(data['Need'].sum() / month_hours, 0)
all_employees = range(1, int(num_employee)+1,1)
#max_employee = round(data['Need'].sum() / norm_hours + 1, 0)
```

```python
#Ввожу все переменные в модель
shedule = {}
for e in all_employees:
    for t in range(data.shape[0]):
        w = data['Week'].iloc[t]
        d = data['Date'].iloc[t]
        t = data['Time'].iloc[t]
        shedule[(e, w, d, t)] = model.NewBoolVar(f"{e}, {w}, {d}, {t}")
```

```python
shedule[1,1,'01.04.2020', '8:00 - 8:30'].GetVarValueMap()
```

```python
#Ввожу ограничение на количество рабочих часов в месяце
for e in all_employees:
    model.Add(sum(shedule[e, w, d, t] for w in pd.unique(data['Week']).tolist() \
                  for d in pd.unique(data[data['Week'] == w]['Date']).tolist() \
                  for t in pd.unique(data[data['Date'] == d]['Time']).tolist()) <= month_hours)
```

```python
#Ввожу ограничение на 40 или 36-часовую рабочую неделю
for e in all_employees:
    for w in pd.unique(data['Week']).tolist():
        model.Add(sum(shedule[e, w, d, t] for d in pd.unique(data[data['Week'] == w]['Date']).tolist() for t in pd.unique(data[data['Date'] == d]['Time']).tolist() ) <= week_hours )
```

```python
#Ввожу ограничение по минимальному количесту рабочих часов сотрудника в день:
for e in all_employees:
    for w in pd.unique(data['Week']).tolist():
        for d in pd.unique(data[data['Week'] == w]['Date']).tolist():
            model.Add(sum(shedule[e, w, d, t] for t in pd.unique(data[data['Date'] == d]['Time']).tolist())>=min_day_time)
```

```python
#Ввожу ограничение по минимальному количесту рабочих часов сотрудника в день:
for e in all_employees:
    for w in pd.unique(data['Week']).tolist():
        for d in pd.unique(data[data['Week'] == w]['Date']).tolist():
            model.Add(sum(shedule[e, w, d, t] for t in pd.unique(data[data['Date'] == d]['Time']).tolist())<=max_day_time)
```

```python
from ortools.sat.python import cp_model


def SimpleSatProgram():
    """Minimal CP-SAT example to showcase calling the solver."""
    # Creates the model.
    model = cp_model.CpModel()

    # Creates the variables.
    num_vals = 3
    x = model.NewIntVar(0, num_vals - 1, 'x')
    y = model.NewIntVar(0, num_vals - 1, 'y')
    z = model.NewIntVar(0, num_vals - 1, 'z')

    # Creates the constraints.
    model.Add(x != y)

    # Creates a solver and solves the model.
    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    if status == cp_model.FEASIBLE:
        print('x = %i' % solver.Value(x))
        print('y = %i' % solver.Value(y))
        print('z = %i' % solver.Value(z))


SimpleSatProgram()
```
